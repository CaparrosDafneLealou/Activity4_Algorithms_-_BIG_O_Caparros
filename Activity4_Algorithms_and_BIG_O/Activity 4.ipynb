```python
# Step 1: Define an algorithm in pseudocode as comments
#Algorithm to fine the largest number in a list

# 1. Assume the first number is the largest
# 2. Compare each number with the current largest
# 3. If a number is larger, update the largest 
# 4. Return the largest number

# Step 2: Implement it in python
numbers = [10, 5, 20, 8, 15]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num

print("Largest number:", largest)

#Try it: for smallest number
smallest = numbers[0]
for num in numbers:
    if num < smallest:
        smallest = num

print("Smallest number:", smallest)
```

```python
# Steo 1: Define a simple linear loop
def print_items(items):
    for item in items:
        print(item)

# Step 2: Observe the number of operations relative to input size
sample = list(range(180))
print_items(sample) # O(n)

#Try it:
#Creating a function for O(1):
def print_first_item(items):
    if items:
        print("First item: {items[0]}")
    else: 
        print("List is empty")

#Creating a function for pairs
def print_pairs(items):
    for i in items:
        for j in items:
            print(f"Pairs: {i},{j}")
sample_data = list(range(180))
print("---testing O(1)---")
print_first_item(sample_data)

print("\n----Testing O(n^2)----")
print_pairs(sample_data)
```

```python
#Step 1: Implement linear search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
#Step 2: Test the fuction
data = [3, 5, 2, 4, 9]
index = linear_search(data, 4)
print("Found at index", index)

#Try now:
def linear_search_with_count(arr, target):
    comparisons = 0 #initial counter
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons
    return -1, comparisons
#Step 2: Test the fuction
data = [3, 5, 2, 4, 9]
target = 4

index, count = linear_search_with_count(data, target)

if index != 1:
    print(f"Target {target} found at index: {index}")
    print(f"Total comparisons made:{count}")
else:
    print("Trget not found.")
```

```python
# Step 1: Implement binary search
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
# Step 2: Test with a sorted list
sorted_data = [1 ,3, 5, 7, 9, 11]
result = binary_search(sorted_data, 7)
print("Found at index:", result)

```

```python
#Step 1: Implement bubble sort
def bubble_sort(arr): 
    n = len(arr)
    for i in range(n): 
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
#Step 2: Test sorting
to_sort = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(to_sort)
print("Sorted list:", to_sort)
```

```python
# Step 1: Implement selection sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
# Step 2: Test and observe perfomance
sample_data = [29, 10, 14, 37, 13]
selection_sort(sample_data)
print("Sorted list:", sample_data)
```

```python
# Step 1: Implement insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
# Step 2: Test and measure performance
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array:", arr)
```

```python
#Number 10
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a 
print(fib_iter(10))
```

```python
#Number 11
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n-1) + fib_rec(n-2)
print(fib_rec(10))
```

```python
#Number 14
def bubble_sort(arr):
    n = len(arr)
    comparisons, swaps = 0, 0
    for i in range(n):
        for j in range(0, n-i-1):
            comparisons += 1
            if arr[j] > arr[j+1]:
                swaps += 1
                arr[j], arr[j+1] = arr[j+1], arr[j]
        return arr, comparisons, swaps
print(bubble_sort([5, 3, 8, 4, 2]))
```

```python
#Number 15
def selection_sort(arr):
    comparisons, swaps = 0, 0 
    for i in range(len(arr)): 
        min_idx = i
        for j in range(i+1, len(arr)):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        swaps += 1
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr, comparisons, swaps
print(selection_sort([5, 3, 8, 4, 2]))
```

```python
#nUMBER 16
def insertion_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            comparisons += 1
            arr[j + 1] = arr[j]
            swaps += 1
            j -= 1
        arr[j + 1] = key
    return arr, comparisons, swaps
print(insertion_sort([5, 3, 8, 4, 2]))
```
